**Exploratory Data Analysis** on the United Nations Global Terrorism Analysis (UNGTA) / Global Terrorism Database (GTD)  
using Python, Pandas, Matplotlib & Seaborn.

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Styling
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
%matplotlib inline

print("✅ All libraries imported successfully!")
```

```python
# ====================== DATA LOADING ======================
df = pd.read_csv('data/globalterrorismdb_0718dist.csv', 
                 encoding='ISO-8859-1', 
                 low_memory=False)

print("✅ Dataset loaded successfully!")
print(f"Shape: {df.shape}")
```

```python
# ====================== MISSING VALUES ======================
drop_cols = [col for col in df.columns if df[col].isnull().mean() > 0.9]
df.drop(columns=drop_cols, inplace=True, errors='ignore')

df['nkill'] = df['nkill'].fillna(0)
df['nwound'] = df['nwound'].fillna(0)
df['total_casualties'] = df['nkill'] + df['nwound']
df['gname'] = df['gname'].fillna('Unknown')
df['city'] = df['city'].fillna('Unknown')

print("✅ Missing values handled!")
```

```python
# ====================== OUTLIER HANDLING ======================
def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    upper = Q3 + 3 * IQR
    return series.clip(upper=upper)

df['nkill_capped'] = cap_outliers(df['nkill'])
df['nwound_capped'] = cap_outliers(df['nwound'])

print("✅ Outliers capped using 3*IQR method.")
```

```python
# ====================== VISUALIZATIONS ======================
# (Sab 6 visualizations yahan hain - pehle wale cells se same)
plt.figure(figsize=(12,6))
attacks_year = df.groupby('iyear').size()
attacks_year.plot(kind='line', marker='o', color='red', linewidth=2)
plt.title('Number of Terrorist Attacks per Year (1970–2017)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.show()
```

## 📊 Key Insights

- Global terrorist attacks mein 2000 ke baad bahut tez increase hua, sabse zyada 2014 mein hue.
- Sabse zyada affected countries: Iraq, Pakistan, India aur Afghanistan.
- Bombing/Explosion sabse common attack type hai (lagbhag 50%).
- Middle East aur South Asia mein sabse zyada incidents hue.
- Unknown groups purane data mein zyada hain, recent years mein named groups (Taliban, ISIS, Boko Haram) dominate kar rahe hain.
