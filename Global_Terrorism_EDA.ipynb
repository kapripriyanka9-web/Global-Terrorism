**Exploratory Data Analysis** on the United Nations Global Terrorism Analysis (UNGTA) / Global Terrorism Database (GTD)  
using Python, Pandas, Matplotlib & Seaborn.

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Styling
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
%matplotlib inline

print("â All libraries imported successfully!")
```

```python
# ====================== DATA LOADING ======================
df = pd.read_csv('data/globalterrorismdb_0718dist.csv', 
                 encoding='ISO-8859-1', 
                 low_memory=False)

print("â Dataset loaded successfully!")
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()[:15]} ...")
df.head()
```

```python
# ====================== MISSING VALUES HANDLING ======================
print("\nð Missing Values Percentage (Top 10 columns):")
missing = (df.isnull().sum() / len(df) * 100).sort_values(ascending=False).head(10)
print(missing)

# Drop columns with >90% missing
drop_cols = missing[missing > 90].index.tolist()
df.drop(columns=drop_cols, inplace=True, errors='ignore')

# Fill key numerical columns
df['nkill'] = df['nkill'].fillna(0)
df['nwound'] = df['nwound'].fillna(0)
df['total_casualties'] = df['nkill'] + df['nwound']

# Fill categorical columns
df['gname'] = df['gname'].fillna('Unknown')
df['city'] = df['city'].fillna('Unknown')

print("\nâ Missing values handled!")
print(df[['nkill', 'nwound', 'total_casualties', 'gname', 'country_txt']].isnull().sum())
```
