using Python, Pandas, Matplotlib & Seaborn.

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Styling
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
%matplotlib inline

print("✅ All libraries imported successfully!")
```

```python
# ====================== DATA LOADING ======================
df = pd.read_csv('data/globalterrorismdb_0718dist.csv', 
                 encoding='ISO-8859-1', 
                 low_memory=False)

print("✅ Dataset loaded successfully!")
print(f"Shape: {df.shape}")
```

```python
# ====================== MISSING VALUES ======================
drop_cols = [col for col in df.columns if df[col].isnull().mean() > 0.9]
df.drop(columns=drop_cols, inplace=True, errors='ignore')

df['nkill'] = df['nkill'].fillna(0)
df['nwound'] = df['nwound'].fillna(0)
df['total_casualties'] = df['nkill'] + df['nwound']
df['gname'] = df['gname'].fillna('Unknown')
df['city'] = df['city'].fillna('Unknown')

print("✅ Missing values handled!")
```

```python
# ====================== OUTLIER HANDLING ======================
def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    upper = Q3 + 3 * IQR
    return series.clip(upper=upper)

df['nkill_capped'] = cap_outliers(df['nkill'])
df['nwound_capped'] = cap_outliers(df['nwound'])

print("✅ Outliers capped using 3*IQR method.")
```

```python
# ====================== VISUALIZATION 1 ======================
plt.figure(figsize=(12,6))
attacks_year = df.groupby('iyear').size()
attacks_year.plot(kind='line', marker='o', color='red', linewidth=2)
plt.title('Number of Terrorist Attacks per Year (1970–2017)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.show()
```

```python
# ====================== VISUALIZATION 2 ======================
plt.figure(figsize=(12,6))
top_countries = df['country_txt'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Countries with Most Terrorist Attacks', fontsize=14)
plt.xlabel('Number of Attacks')
plt.ylabel('Country')
plt.show()
```

```python
# ====================== VISUALIZATION 3 ======================
plt.figure(figsize=(10,8))
attack_types = df['attacktype1_txt'].value_counts().head(8)
plt.pie(attack_types.values, labels=attack_types.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Attack Types', fontsize=14)
plt.axis('equal')
plt.show()
```

```python
# ====================== VISUALIZATION 4 ======================
plt.figure(figsize=(10,6))
region_counts = df['region_txt'].value_counts()
sns.barplot(x=region_counts.values, y=region_counts.index, palette='magma')
plt.title('Terrorist Attacks by Region', fontsize=14)
plt.xlabel('Number of Attacks')
plt.ylabel('Region')
plt.show()
```

```python
# ====================== VISUALIZATION 5 ======================
plt.figure(figsize=(12,6))
top_groups = df['gname'].value_counts().head(10)
sns.barplot(x=top_groups.values, y=top_groups.index, palette='coolwarm')
plt.title('Top 10 Perpetrator Groups', fontsize=14)
plt.xlabel('Number of Attacks')
plt.ylabel('Group Name')
plt.show()
```

```python
# ====================== VISUALIZATION 6 ======================
plt.figure(figsize=(12,6))
casualties_year = df.groupby('iyear')['total_casualties'].sum()
casualties_year.plot(kind='bar', color='darkred')
plt.title('Total Casualties (Killed + Wounded) per Year', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Total Casualties')
plt.show()
```

## 📊 Key Insights

- Global terrorist attacks increased dramatically after 2000 and peaked around 2014.
- Iraq, Pakistan, India, and Afghanistan are the most affected countries.
- Bombing/Explosion is the most common attack type (~50%).
- Middle East & South Asia together account for more than 60% of all attacks.
- Unknown groups dominate early years while named groups (Taliban, ISIS, Boko Haram etc.) dominate recent years.
- Highest casualty years coincide with major conflicts (Iraq, Syria, Afghanistan).
